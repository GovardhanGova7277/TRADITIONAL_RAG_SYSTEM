```python
## Create a txt fule

import os
os.makedirs('../data/text_files',exist_ok=True)
```

```python
## Adding the text content to python.txt file

rag_demo_data = {
    "../data/text_files/python_file.txt": (
        "Retrieval-Augmented Generation (RAG) is a technique that grants LLMs "
        "access to specific, private data. It works by converting text into "
        "vectors and using Cosine Similarity to find the most relevant pieces "
        " of information to answer a user's question."
    ),
    "../data/text_files/vector_basics.txt": (
        "Vector databases like ChromaDB or FAISS store mathematical representations "
        "of text. When a user asks a query, the system calculates the distance "
        "between the query vector and the stored document vectors."
    ),
    "../data/text_files/deployment_notes.txt": (
        "For a zero-rupee deployment, one can use Ollama for local LLM inference "
        "and LangChain for orchestration. This avoids expensive API costs while "
        "keeping data private on a local machine."
    )
}

for filepath, content in rag_demo_data.items():
    with open(filepath,mode='w',encoding="utf-8") as f:
        f.write(content.strip())
        
print("Sample text file is created")
    
```

```python
### Textloader

from langchain_community.document_loaders import TextLoader
```

```python
## read the text

loader = TextLoader("../data/text_files/python_file.txt",encoding="utf-8")

document = loader.load()

for some in document:
    print(some,"\n")
```

```python
## Directory Loader

from langchain_community.document_loaders import DirectoryLoader

dir_loader = DirectoryLoader(
    "../data/text_files",
    glob="**/*.txt",
    loader_cls=TextLoader,
    loader_kwargs={'encoding':'utf-8'},
    show_progress=False
)
```

```python
dir_loader.load()
```

```python
## Loading PDFs
from langchain_community.document_loaders import PyPDFLoader,PyMuPDFLoader

dir_loader = DirectoryLoader(
    "../data/pdf",
    glob="**/*.pdf",
    loader_cls=PyMuPDFLoader, # Class to use
    show_progress=False
)

pdf_documents = dir_loader.load()
pdf_documents
```
